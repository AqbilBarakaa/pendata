# Diskritisasi data (Binning) menggunakan K-Means Clustering 

## Mengganti Label Sepal Length menjadi Kategori dengan 4 Cluster

### Import Library

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

### Mengambil Data IRIS

```python
# Baca data
data = pd.read_csv("IRIS_Diskritisasi.csv")

print("Data IRIS")
print(data.head())
```

### Mengambil label Sepal Length

```python
# Ambil hanya kolom sepal_length untuk clustering
X = data[['sepal_length']].values

print("\nData IRIS Sepal Length")
print(X)
```

### Melakukan Normalisasi Data

```python
# Normalisasi
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("\nNormalisasi data Sepal Length")
print(X_scaled)
```

### Melakukan Clustering dengan 4 Cluster

```python
# Clustering
kms = KMeans(n_clusters=4, random_state=0, n_init="auto").fit(X_scaled)
print("\nHasil label cluster (dengan angka):")
print(kms.labels_)
```

### Mengubah Label Cluster menjadi angka

```python
# Mapping label angka → huruf
label_mapping = {0:'A', 1:'B', 2:'C', 3:'D'}
cluster_labels_named = [label_mapping[label] for label in kms.labels_]
cluster_labels_named_np = np.array(cluster_labels_named)

# Cetak hasil label dalam bentuk huruf
print("\nHasil label cluster (dengan nama):")
print(cluster_labels_named_np)
```

### Mengubah Data Sepal Length menjadi Cluster yang telah dilakukan sebelumnya dan disimpan ke CSV

```python
# Ganti kolom sepal_length dengan label huruf cluster
data['sepal_length'] = cluster_labels_named_np
print("\nData IRIS dengan 'sepal_length' diganti huruf cluster:")
print(data)

# Simpan ke CSV
data.to_csv("IRIS_Diskritisasi.csv", index=False)
```
