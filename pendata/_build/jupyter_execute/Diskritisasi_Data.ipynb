## Melakukan Diskritisasi data menggunakan K-Means Clustering

### Mengganti Label Sepal Length menjadi Kategori dengan 4 Cluster

Import Library

```python
%pip install scikit-learn


import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

Mengambil Data IRIS

```python
# Baca data
data = pd.read_csv("IRIS_Diskritisasi.csv")

print("Data IRIS")
print(data.head())
```

 Mengambil label Sepal Length

```python
# Ambil hanya kolom sepal_length untuk clustering
X = data[['sepal_length']].values

print("\nData IRIS Sepal Length")
print(X)
```

```python
# Clustering
kms = KMeans(n_clusters=4, random_state=0, n_init="auto").fit(X_scaled)
print("\nHasil label cluster (dengan angka):")
print(kms.labels_)
```

```python
# Ganti kolom sepal_length dengan label huruf cluster
data['sepal_length'] = cluster_labels_named_np
print("\nData IRIS dengan 'sepal_length' diganti huruf cluster:")
print(data)

# Simpan ke CSV
data.to_csv("IRIS_Diskritisasi.csv", index=False)
```

 Import Library

```python
# Import Library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
```

### Implementasi Menggunakan Outlier Removal (penghapusan manual berdasarkan sepal_width)

Mengambil data yang sudah di Diskritisasi

```python
# 1. Baca data
dataset = pd.read_csv('IRIS_Diskritisasi.csv')
print("=== DATA AWAL SEBELUM HAPUS OUTLIER ===")
dataset
```

Menghapus data yang dianggap outlier (di sini: kategori 'A' pada sepal_width).

```python
# Penghapusan outlier: buang baris dengan kategori ekstrem (misal: 'A' pada sepal_width)
dataset_outlier_removed = dataset[dataset['sepal_width'] != 'A']

print("Data Setelah Outlier Removal:")
dataset_outlier_removed
```

```python
# 4. Split data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=43)
```

Membuat dan melatih model Naive Bayes khusus untuk data kategori.

```python
# 5. Model Naive Bayes untuk data kategori
classifier = CategoricalNB()
classifier.fit(X_train, y_train)
```

Melakukan prediksi terhadap data uji.

```python
# 6. Prediksi
y_pred = classifier.predict(X_test)
y_pred
```

Mengukur akurasi dan menampilkan confusion matrix.

```python
# 7. Evaluasi
print("\n[VERSI DENGAN OUTLIER REMOVAL (buang sepal_width=='A')]")
print("Accuracy:", accuracy_score(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

## Melakukan Klasifikasi dengan model Naive Bayas menggunakan data yang belum di Diskritisasi

### Implementasi Dengan Outlier Removal

Import Library.

```python
# Import Library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
```

Mengambil data IRIS asli.

```python
# 1. Baca data
datairis = pd.read_csv('iris_dataset.csv')
datairis
```

Menghapus data ekstrem (outlier) pada kolom sepal_width menggunakan metode IQR.

```python
# Outlier removal berdasarkan kolom numerik 'sepal_width' (IQR method)
Q1 = datairis['sepal_width'].quantile(0.25)
Q3 = datairis['sepal_width'].quantile(0.75)
IQR = Q3 - Q1
datairis = datairis[(datairis['sepal_width'] >= Q1 - 1.5 * IQR) & (datairis['sepal_width'] <= Q3 + 1.5 * IQR)]

print("Data Setelah Outlier Removal:")
datairis
```

Memisahkan Fitur dan Label

```python
# 2. Ambil fitur dan label
X = datairis.iloc[:, 1:5].values
y = datairis['class'].values
```

Mengubah label kelas dari teks ke angka.

```python
# 3. Encode label class
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

Pisah Dataset ke Data Latih dan Data Uji.

```python
# 4. Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=43)
```

Membuat dan melatih model Naive Bayes.

```python
# 5. Model Naive Bayes Gaussian
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

Melakukan prediksi terhadap data uji.

```python
# 6. Prediksi
y_pred = classifier.predict(X_test)
y_pred
```

Mengukur akurasi dan membuat matriks kebingungan.

```python
# 7. Evaluasi
print("\n[VERSI DENGAN OUTLIER REMOVAL]")
print("Accuracy:", accuracy_score(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

Menampilkan nilai asli dan prediksi dalam bentuk tabel.

```python
# 8. Tampilkan hasil prediksi
df = pd.DataFrame({
    'Nilai Asli': le.inverse_transform(y_test),
    'Nilai Prediksi': le.inverse_transform(y_pred)
})
print("\nHasil prediksi:")
df
```

## Melakukan Klasifikasi dengan model Pohon Keputusan menggunakan data yang sudah di Diskritisasi

Import Library

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

Mengambil data diskritisasi

```python
# 1. Baca data diskritisasi
df = pd.read_csv('IRIS_Diskritisasi.csv')
```

Encode label kelas ke bentuk numerik.

```python
# 2. Encode label kelas
le = LabelEncoder()
df['ClassEncoded'] = le.fit_transform(df['class'])
```

Tentukan daftar fitur kategorikal yang digunakan.

```python
# 3. Fitur kategorikal
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

results = {}
thresholds = {}
```

Hitung Information Gain tiap fitur dan simpan threshold terbaiknya.

```python
# 4. Hitung Information Gain & simpan threshold terbaik tiap fitur
for feature in features:
    unique_values = sorted(df[feature].unique())  # Contoh: ['A', 'B', 'C', 'D']

    best_ig = -1
    best_threshold = None
    best_entropies = {}

    for threshold in unique_values:
        df[f'{feature}_binary'] = df[feature].apply(lambda x: 1 if x > threshold else 0)

        X = df[[f'{feature}_binary']]
        y = df['ClassEncoded']

        clf = DecisionTreeClassifier(criterion='entropy', max_depth=1, random_state=42)
        clf.fit(X, y)

        tree = clf.tree_
        if tree.node_count < 3:
            continue

        n_total = tree.n_node_samples[0]
        left_count = tree.n_node_samples[1]
        right_count = tree.n_node_samples[2]

        e_root = tree.impurity[0]
        e_left = tree.impurity[1]
        e_right = tree.impurity[2]

        avg_entropy = (left_count / n_total) * e_left + (right_count / n_total) * e_right
        info_gain = e_root - avg_entropy

        if info_gain > best_ig:
            best_ig = info_gain
            best_threshold = threshold
            best_entropies = {
                'Entropy Awal': round(e_root, 4),
                'Entropy Kiri': round(e_left, 4),
                'Entropy Kanan': round(e_right, 4),
                'Entropy Rata2': round(avg_entropy, 4),
                'Information Gain': round(info_gain, 4)
            }

    results[feature] = {
        'Threshold Terbaik': best_threshold,
        **best_entropies
    }
    thresholds[feature] = best_threshold
    df[f'{feature}_binary'] = df[feature].apply(lambda x: 1 if x > best_threshold else 0)
```

Tampilkan hasil IG dan tentukan fitur dengan IG tertinggi.

```python
# 5. Tampilkan hasil IG
print("=== Hasil Information Gain (Fitur Diskritisasi) ===\n")
best_feature = None
best_ig_value = -1

for feature, info in results.items():
    print(f"Fitur: {feature}")
    print(f"  Threshold Terbaik : {info['Threshold Terbaik']}")
    print(f"  Entropy Awal      : {info['Entropy Awal']}")
    print(f"  Entropy Kiri      : {info['Entropy Kiri']}")
    print(f"  Entropy Kanan     : {info['Entropy Kanan']}")
    print(f"  Entropy Rata-rata : {info['Entropy Rata2']}")
    print(f"  Information Gain  : {info['Information Gain']}\n")

    if info['Information Gain'] > best_ig_value:
        best_ig_value = info['Information Gain']
        best_feature = feature
        best_threshold = info['Threshold Terbaik']

print(f"Fitur terbaik berdasarkan Information Gain tertinggi: **{best_feature}** (IG = {best_ig_value})")
```

Latih model dengan semua fitur biner hasil threshold terbaik.

```python
# 6. Training model dengan semua fitur biner
X = df[[f'{f}_binary' for f in features]]
y = df['ClassEncoded']
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
clf.fit(X, y)
```

Visualisasikan pohon keputusan (maksimal kedalaman 3).

```python
# 7. Visualisasi pohon keputusan
plt.figure(figsize=(10, 6))
plot_tree(clf,
          feature_names=[f'{f}_binary' for f in features],
          class_names=le.classes_,
          filled=True,
          rounded=True)
plt.title("Visualisasi Pohon Keputusan (Semua Fitur)")
plt.tight_layout()
plt.show()
```

Tampilkan aturan hasil pohon keputusan.

```python
# 8. Aturan dari Decision Tree
print("\n=== Aturan dari Decision Tree ===\n")
rules = export_text(clf, feature_names=[f'{f}_binary' for f in features])
print(rules)
```

Prediksi 2 data baru dengan model yang telah dibuat.

```python
# 9. Uji dengan 2 data baru (diskrit)
print("=== Hasil Prediksi 2 Data Baru ===\n")

# Dua data baru
new_data = pd.DataFrame({
    'sepal_length': ['C', 'A'],
    'sepal_width': ['B', 'D'],
    'petal_length': ['B', 'C'],
    'petal_width': ['A', 'C']
})

# Binarisasi semua fitur sesuai threshold masing-masing
for f in features:
    t = thresholds[f]
    new_data[f'{f}_binary'] = new_data[f].apply(lambda x: 1 if x > t else 0)

# Prediksi
X_new = new_data[[f'{f}_binary' for f in features]]
predictions = clf.predict(X_new)
predicted_labels = le.inverse_transform(predictions)

# Tampilkan hasil prediksi
for i, label in enumerate(predicted_labels):
    print(f"Data ke-{i+1}:")
    for f in features:
        print(f"  {f.replace('_', ' ').title()} = {new_data.iloc[i][f]}")
    print(f"  Prediksi Kelas: {label}\n")
```

## Melakukan Klasifikasi dengan model Pohon Keputusan menggunakan data yang belum di Diskritisasi

Mengambil data IRIS asli

```python
# 1. Load dataset iris (belum didiskritisasi)
df = pd.read_csv('iris_dataset.csv')  # Pastikan kolom: sepal_length, sepal_width, petal_length, petal_width, class
```

Melakukan encoding pada label kelas (class) menjadi numerik dengan LabelEncoder.

```python
# 2. Encode class target
le = LabelEncoder()
df['ClassEncoded'] = le.fit_transform(df['class'])
```

Menyimpan daftar nama fitur numerik yang akan dianalisis.

```python
# 3. Fitur numerik
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

results = {}
```

Menghitung Information Gain (IG) dan mencari threshold terbaik untuk masing-masing fitur menggunakan pohon keputusan sederhana.

```python
# 4. Hitung IG dan cari threshold terbaik per fitur
for feature in features:
    thresholds = sorted(df[feature].unique())
    best_ig = -1
    best_threshold = None
    best_detail = {}

    for threshold in thresholds:
        df['temp_binary'] = (df[feature] > threshold).astype(int)
        X = df[['temp_binary']]
        y = df['ClassEncoded']

        clf = DecisionTreeClassifier(criterion='entropy', max_depth=1, random_state=0)
        clf.fit(X, y)
        tree = clf.tree_

        if tree.node_count < 3:
            continue

        n_total = tree.n_node_samples[0]
        left_count = tree.n_node_samples[1]
        right_count = tree.n_node_samples[2]

        e_root = tree.impurity[0]
        e_left = tree.impurity[1]
        e_right = tree.impurity[2]

        avg_entropy = (left_count / n_total) * e_left + (right_count / n_total) * e_right
        ig = e_root - avg_entropy

        if ig > best_ig:
            best_ig = ig
            best_threshold = threshold
            best_detail = {
                'Entropy Awal': round(e_root, 4),
                'Entropy Kiri': round(e_left, 4),
                'Entropy Kanan': round(e_right, 4),
                'Entropy Rata2': round(avg_entropy, 4),
                'Information Gain': round(ig, 4)
            }

    results[feature] = {
        'Threshold Terbaik': best_threshold,
        **best_detail
    }
```

Menampilkan hasil IG dan threshold terbaik dari setiap fitur. Sekaligus menentukan fitur dengan IG tertinggi.

```python
# 5. Tampilkan IG untuk semua fitur
print("=== Hasil Information Gain per Fitur ===\n")
best_feature = None
best_ig_value = -1

for feature, info in results.items():
    print(f"Fitur: {feature}")
    print(f"  Threshold Terbaik : {info['Threshold Terbaik']}")
    print(f"  Entropy Awal      : {info['Entropy Awal']}")
    print(f"  Entropy Kiri      : {info['Entropy Kiri']}")
    print(f"  Entropy Kanan     : {info['Entropy Kanan']}")
    print(f"  Entropy Rata-rata : {info['Entropy Rata2']}")
    print(f"  Information Gain  : {info['Information Gain']}\n")

    if info['Information Gain'] > best_ig_value:
        best_ig_value = info['Information Gain']
        best_feature = feature
        best_threshold = info['Threshold Terbaik']

print(f"Fitur terbaik berdasarkan Information Gain tertinggi: **{best_feature}** (IG = {best_ig_value})\n")
```

Membuat kolom binary berdasarkan fitur terbaik dan threshold-nya (untuk keperluan analisis jika dibutuhkan).

```python
# 6. Tambahkan kolom binary dari fitur terbaik
df[f'{best_feature}_binary'] = (df[best_feature] > best_threshold).astype(int)
```

Melatih model Decision Tree penuh dengan semua fitur numerik (bukan biner), menggunakan kedalaman maksimal 3.

```python
# 7. Gunakan semua fitur asli (numerik) untuk training pohon lengkap
X = df[features]
y = df['ClassEncoded']
clf_full = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)  # lebih dari 3 node
clf_full.fit(X, y)
```
